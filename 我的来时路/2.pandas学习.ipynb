# Series

```python
s = pd.Series([1,2,3,4,5])
print(s)
s = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
print(s)
s = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'], name='zzz')
print(s)
```

```python
# 通过字典创建
s = pd.Series({'a':1,'b':2,'c':3,'d':4})
print(s)
s2 = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
s1 = pd.Series(s2, index=['a','c','d'])
print(s1)
```

```python
print(s.dtype)
print(s.index)
```

```python
print(s.loc["a":"c"])
print(s.iloc[0])
print(s.at["a"])
print(s.iat[0])
```

```python
print(s["a"])
print(s[s<3])
```

```python
s = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
s.head()
s.tail()
s.describe()
```

```python
import numpy as np
np.random.seed(42)
index = []
for i in range(10):
    index.append(i)
scores = pd.Series(np.random.randint(50,101,10),index)
print(scores)
scores.describe()
```

# DataFrame

```python
import pandas as pd
import numpy as np
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,10])
df = pd.DataFrame({"第1列":s1,"第2列":s2})
print(df)
print(type(df))
```

```python
df = pd.DataFrame(
    {
        "name":["tom","jack","alice","bob"],
        "age":[22,22,22,22],
        "score":[1,2,3,4]
    },index=[1,2,3,4],columns=["name","age","score"]
)
print(df)
```

```python
#属性
print(df.index)
print(df.columns)
print(df.values)
```

```python
print(df.ndim)
print(df.shape)
print(df.dtypes)
print(df.size)
print(df.T)
print(df.T.index)
```

```python
print(df.loc[4])
print(df.iloc[3])   #隐式
```

```python
print(df.loc[:,"score"])
print(df.iloc[:,0])
```

```python
print(df.at[1,"score"])
print(df.iat[2,1])
```
